```python
#creates data for a heatmap with x and y axes of core-mantle density ratio and neutrino energy
#then returns four different lists of data that can be used as the heatmaps "z" axis
def make_density_ratio_vs_energy_heatmap_data():
    energies = np.linspace(1, 10, 10)#[1, 2, 3, 5, 7, 10, 20, 50, 75, 100]
    ratios = np.linspace(0, 10, 10)
    elec_probabilities = []
    muon_probabilities = []
    tau_probabilities = []
    muon_to_electron_ratios = []
    for energy in energies:
        for ratio in ratios:
            new_probabilities = monte_carlo_uniform_core(energy, ratio, 330)
            elec_probabilities.append(new_probabilities[0])
            muon_probabilities.append(new_probabilities[1])
            tau_probabilities.append(new_probabilities[2])
            muon_to_electron_ratios.append(new_probabilities[1] / new_probabilities[0])

    return elec_probabilities, muon_probabilities, tau_probabilities, muon_to_electron_ratios
```

```python
#creates a heatmap to display the muon probabilities for a given energy and core-mantle density ratio
#takes data from make_density_ratio_vs_energy_heatmap_data
def visualize_density_ratio_and_energy_heatmap(flavor_ratios):
    """energies = np.linspace(1, 10, 100)#[1, 2, 3, 5, 7, 10, 20, 50, 75, 100]
    ratios = np.linspace(1, 100, 100)#[1, 2, 3, 5, 7, 10, 20, 50, 75, 100]
    """
    
    energies = np.linspace(1, 10, 10)#[1, 2, 3, 5, 7, 10, 20, 50, 75, 100]
    ratios = np.linspace(0, 10, 10)
    
    # Create a meshgrid
    X, Y = np.meshgrid(ratios, energies)     
    # Plot the heatmap
    plt.figure(figsize=(8, 6))
    plt.pcolormesh(X, Y, flavor_ratios, shading='auto', cmap='plasma')
    plt.colorbar(label='Muon-Electron Flavor Ratio')
    plt.xlabel('Core-Mantle Density Ratio')
    plt.ylabel('Energies (GeV)')
    plt.title('Flavor Ratio by Energy and Core-Mantle Density Ratio (R_core = 330 km)') 
    plt.savefig("Density_Ratio_and_energy_heatmap",dpi=300,bbox_inches='tight')
```

```python
#creates data for comparing expected flavor ratio and neutrino energy
def generate_flavor_ratio_vs_energy_data(core_mantle_density_ratio, r_core):
    energies = np.linspace(1, 10, 100)
    flavor_ratios = [] # ELECTRON TO MUON RATIO
    for energy in energies:
        simulated_flavors = monte_carlo_uniform_core(energy, core_mantle_density_ratio, r_core)
        flavor_ratios.append(simulated_flavors[1] / simulated_flavors[0])
    
    return energies, flavor_ratios
```

```python
#make some data to graph flavor ratios for different core densities
ratio1 = generate_flavor_ratio_vs_energy_data(1, 330) #3.5 g/cm^3
ratio2 = generate_flavor_ratio_vs_energy_data(2, 330) # 7 g/cm^3
ratio3 = generate_flavor_ratio_vs_energy_data(9.1, 330) # 30 g/cm^3
ratio4 = generate_flavor_ratio_vs_energy_data(36, 330)  # 100 g/cm^3
```

```python

plt.plot(ratio1[0],ratio1[1],label=r'$P_{core} = 3.5 g/cm^3$')
plt.plot(ratio2[0],ratio2[1],label=r'$P_{core} = 7 g/cm^3$')
plt.plot(ratio3[0],ratio3[1],label=r'$P_{core} = 30 g/cm^3$')
plt.plot(ratio4[0],ratio4[1],label=r'$P_{core} = 100 g/cm^3$')


plt.legend(loc='best')
plt.xlabel('Energy (GeV)')
plt.ylabel('Flavor Ratio (muon : electron)')
plt.title('Flavor Ratio vs Energy')
plt.savefig("ratio_vs_energy_for_a_few_densities",dpi=300,bbox_inches='tight')
```
